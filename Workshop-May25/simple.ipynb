```python
import numpy as np
import pandas as pd

np.random.seed(42)

area = np.random.randint(500, 3000, size=10)  
slope = 150  # Price per square foot
intercept = 50000  # Base price
noise = np.random.randint(-10000, 10000, size=10) 

price = slope * area + intercept + noise

data = pd.DataFrame({
    'Area': area,
    'Price': price
})

data 
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Area'], data['Price'], color='blue', label='Data Points')
plt.title('House Area vs Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
np.random.seed(42)

age = np.linspace(0, 50, 10)  # Age of the house in years
a = 200  # Controls the steepness of the curve
h = 25   # Age at which price peaks
k = 300000  # Peak price
noise = np.random.randint(-5000, 5000, size=10)  # Random noise

# Calculate price using the inverted U-shaped equation
price = -a * (age - h)**2 + k + noise

data = pd.DataFrame({
    'Age': age,
    'Price': price
})

print(data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'], data['Price'], color='blue', label='Data Points')
plt.title('Age vs Price')
plt.xlabel('Age')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
